```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Step 1: Specify the folder containing your CSV files
folder_path = "data_original"  # Replace with your folder path

# Step 2: Load all CSV files and extract 'Open' prices
stock_data = {}
for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        # Assuming each CSV has 'Date' and 'Open' columns
        df = pd.read_csv(os.path.join(folder_path, file))
        stock_name = file.replace(".csv", "")  # Use filename as stock identifier
        stock_data[stock_name] = df["Open"]

# Step 3: Combine into a single DataFrame
combined_df = pd.DataFrame(stock_data)

# Step 4: Calculate the correlation matrix
correlation_matrix = combined_df.corr()

# Step 5: Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, center=0)
plt.title("Stock Price Correlation Heatmap (Opening Prices)")
plt.show()
```
